```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize
from sklearn.preprocessing import PolynomialFeatures
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegressionCV
from matplotlib.colors import ListedColormap
%matplotlib inline
```

```python
def plot2data(X, y, lb_x, lb_y, t1, t2, axes=None):
    c1 = y == 0
    c2 = y == 1
    C1 = X[c1]
    C2 = X[c2]
    if axes == None:
        axes = plt.gca()
    axes.scatter(C1[:,0], C1[:,1], marker='o', c='r', s=60, label=t1)
    axes.scatter(C2[:,0], C2[:,1], marker='o', c='g', s=60, label=t2)
    axes.set_xlabel(lb_x)
    axes.set_ylabel(lb_y)
    
def plot3data(X, y, lb_x, lb_y, t1, t2, t3, axes=None):
    c1 = y == 0
    c2 = y == 1
    c3 = y == 2
    C1 = X[c1]
    C2 = X[c2]
    C3 = X[c3]
    if axes == None:
        axes = plt.gca()
    axes.scatter(C1[:,0], C1[:,1], marker='o', c='r', s=60, label=t1)
    axes.scatter(C2[:,0], C2[:,1], marker='o', c='g', s=60, label=t2)
    axes.scatter(C3[:,0], C3[:,1], marker='o', c='b', s=60, label=t3)
    axes.set_xlabel(lb_x)
    axes.set_ylabel(lb_y)

def plotDB(model, X, axes=None):
    x_min, x_max = X[:,0].min(), X[:,0].max()
    y_min, y_max = X[:,1].min(), X[:,1].max()
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    X_G = np.c_[xx.ravel(), yy.ravel()]
    A1, A2 = fc(model, X_G)
    zz = np.argmax(A2, axis=1)
    zz = zz.reshape(xx.shape)
    if axes == None:
        axes = plt.gca()
    axes.contour(xx, yy, zz, linwidth=2, color='g')
    color = ListedColorMap('#FFAAAA', '#AAFFAA', '#AAAAFF')
    axes.pcolormesh(xx, yy, zz, cmap=color)
    plot3data(X, y, "Es1", "Es2", "c1", "c2", "c3")
    
def sigmoid(Z):
     y = y.reshape(-1, 1)
     h = 1.0 / (1.0 + np.exp(-Z))
     return h
    
def softmax(Z):
    h = np.exp(Z) / np.sum(np.exp(Z), axis=1, keepdims=True)
    return h

def af(Z):
    h = np.tanh(Z)
    return h

def costF(model, X, y, reg):
    #向前计算
    A1, A2 = fc(model, X)
    #计算损失
    y_loss = (1.0 / X.shape[0]) *  np.sum(-np.log(A2[range(X.shape[0]), y]))
    reg_loss = (reg / 2.0) * (np.sum(np.square(model['W1'])) + np.sum(np.square(model['W2'])))
    cost = y_loss + reg_loss
    #返回
    return cost

def fc(model, X):
    W1, b1 , W2, b2 = model['W1'], model['b1'], model['W2'], model['b2']
    Z1 = X.dot(W1) + b1
    A1 = af(Z1)
    Z2 = A1.dot(W2) + b2
    A2 = softmax(Z2)
    return A1, A2

def bp(model, X, reg, ep):
    W1, b1, W2, b2 = model['W1'], model['b1'], model['W2'], model['b2']
    A1, A2 = fc(model, X)
    A2[range(A2.shape[0]), y] -= 1
    
    delta3 = A2
    delta2 = (1 - np.power(A1, 2)) * (delta3.dot(W2.T))
    
    dW2 = (A1.T).dot(delta3) + (reg * W2)
    db2 = np.sum(delta3, axis=0, keepdims=True)
    dW1 = (X.T).dot(delta2) + (reg * W1)
    db1 = np.sum(delta2, axis=0, keepdims=True)
    
    W2 += -ep * dW2
    b2 += -ep * db2
    W1 += -ep * dW1
    b1 += -ep * db1
    
    #更新model
    model = {'W1':W1, 'b1':b1, 'W2':W2, 'b2':b2}
    return 

def init(inputs, outputs, hnums):
    np.random.seed(0)
    W1 = np.random.randn(inputs, hnums)
    b1 = np.zeros((1, hnums))
    W2 = np.random.randn(hnums, outputs)
    b2 = np.zeros((1, outputs))
    model = {'W1':W1, 'b1':b1, 'W2':W2, 'b2':b2}
    return model

def train(model, X):
    circlenum = 20000
    for i in xrange(0, circlenum):
        A1, A2 = fc(model, X)
        bp(model, X, 0.1, 0.01)
        if i % 1000 == 0:
            print "Loss in %f is %f " % (i, costF(model, X, y, 0.1))
    return 

def predict(model, X):
    A1, A2 = fc(model, X)
    y_bar = np.argmax(A2, axis=1)
    return y_bar
```

```python
N = 100 # 每类样本个数
D = 2 # 维度
K = 3 # 类别数
X = np.zeros((N*K,D)) # data matrix (each row = single example)
y = np.zeros(N*K, dtype='uint8') # class labels
for j in xrange(K):
    ix = range(N*j,N*(j+1))
    r = np.linspace(0.0,1,N) # 半径
    t = np.linspace(j*4,(j+1)*4,N) + np.random.randn(N)*0.2 # 角度，加上是一个随机噪声
    X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
    y[ix] = j
# lets visualize the data:
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
```

```python
model = init(2, 5, 3)
train(model, X)
y_bar = predict(model, X)
acc = 100.0 * sum(y_bar == y) / y.size
print "Accuracy {}".format(acc, decimals=2)
```
