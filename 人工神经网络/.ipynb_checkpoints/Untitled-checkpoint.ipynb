```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
%matplotlib inline
```

```python
class ANN:
    inputs = np.inf
    outputs = np.inf
    hnns = np.inf
    reg = np.inf
    ep = np.inf
    W1 = np.inf
    b1 = np.inf
    W2 = np.inf
    b2 = np.inf
    A1 = np.inf
    A2 = np.inf
    acc = np.inf
    cost = np.inf
    
    def __init__(self, inputs, outputs, hnns, reg, ep):
        self.inputs = inputs
        self.outputs = outputs
        self.hnns = hnns
        self.reg = reg
        self.ep = ep
        self.W1 = np.random.randn(inputs, hnns)
        self.W2 = np.random.randn(hnns, outputs)
        self.b1 = np.zeros((1, hnns))
        self.b2 = np.zeros((1, outputs))
        
    def train(self, X, y, loops):
        for i in xrange(loops):
            self.fc(X, y)
            self.bp(X, y)
            if i % 1000 == 0:
                print self.costF(X, y)
                
    def fc(self, X, y):
        Z1 = X.dot(self.W1) + self.b1
        self.A1 = self.tanh(Z1)
        Z2 = self.A1.dot(self.W2) + self.b2
        self.A2 = self.softmax(Z2)
    
    def bp(self, X, y):
        self.A2[range(X.shape[0]), y] -= 1
        delta3 = self.A2
        delta2 = delta3.dot(self.W2.T) * ( 1 - np.power(self.A1, 2))
                
        dW2 = (self.A1.T).dot(delta3) + (self.reg * self.W2)
        db2 = np.sum(delta3, axis=0, keepdims=True)
        dW1 = (X.T).dot(delta2) + (self.reg * self.W1)
        db1 = np.sum(delta2, axis=0, keepdims=True)
        self.W2 += -self.ep * dW2
        self.b2 += -self.ep * db2
        self.W1 += -self.ep * dW1
        self.b1 += -self.ep * db1
        
    def costF(self, X, y):
        self.fc(X, y)
        y_bar = self.A2[range(X.shape[0]), y]
        y_loss = (1.0 / X.shape[0]) * np.sum(-np.log(y_bar))
        r_loss = (self.reg / 2.0) * (np.sum(np.square(self.W1)) + np.sum(np.square(self.W2)))
        loss = y_loss + r_loss
        return loss
    
    def tanh(self, X):
        A_1 = np.tanh(X)
        return A_1
    
    def softmax(self, X):
        A_2 = np.exp(X) / np.sum(np.exp(X), axis=1, keepdims=True)
        return A_2
      
    def predict(self, X):
        self.fc(X, y)
        pt = np.argmax(self.A2, axis=1)
        return  pt
    
    def plotdb(self, X, y, axes=None):
        x_min, x_max = X[:,0].min() - 0.5, X[:,0].max() + 0.5
        y_min, y_max = X[:,1].min() - 0.5, X[:,1].max() + 0.5
        xx, yy = np.meshgrid(np.linspace(x_min, y_max, 500), np.linspace(y_min, y_max, 500))
        X_G = np.c_[xx.ravel(), yy.ravel()]
        self.fc(X_G, y)
        zz = np.argmax(self.A2, axis=1)
        zz = zz.reshape(xx.shape)
        color = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
        if axes == None:
            axes = plt.gca()
        plt.pcolormesh(xx, yy, zz, cmap=color)
        plt.contour(xx, yy, zz, color='g')
        plt.scatter(X[:,0], X[:,1], s=20, c=y, cmap=plt.cm.spectral)
        
    def loaddata(self):
        N = 100 # 每类样本个数
        D = 2 # 维度
        K = 3 # 类别数
        X = np.zeros((N*K,D)) # data matrix (each row = single example)
        y = np.zeros(N*K, dtype='uint8') # class labels
        for j in xrange(K):
            ix = range(N*j,N*(j+1))
            r = np.linspace(0.0,1,N) # 半径
            t = np.linspace(j*4,(j+1)*4,N) + np.random.randn(N)*0.2 # 角度，加上是一个随机噪声
            X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
            y[ix] = j
        # lets visualize the data:
        #plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
        return X, y

    def plotSd(self, X, y):
        plt.figure(figsize=(16, 32))
        h_dims = [1, 2, 4, 16, 32, 64, 128, 256, 512, 1024]
        for i,dim in enumerate(h_dims):
            plt.subplot(5,3,i+1)
            plt.title("Hidden layer size {}".format(dim))
            self.hnns = h_dims
            self.train(X, y, 20000)
            self.plotdb(X, y)    
```

```python
net = ANN(2, 3, 10, 0.01, 0.01)
X, y = net.loaddata()
net.train(X, y, 10000)
net.plotdb(X, y)
y_p = net.predict(X)
print "Acc {}".format(100.0 * sum(y_p == y) / y.size)
```
