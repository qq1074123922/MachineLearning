```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize
from sklearn.preprocessing import PolynomialFeatures
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegressionCV
from matplotlib.colors import ListedColormap
%matplotlib inline
```

```python
#绘制2分类散点图
def plotdata(X, y, lb_x, lb_y, t_1, t_2, axes=None):
    c1 = np.in1d(y, [0])
    c2 = np.in1d(y, [1])
    c1 = y == 0
    c2 = y == 1
    C1 = X[c1]
    C2 = X[c2]
    if axes == None:
        axes = plt.gca()
    axes.scatter(C1[:,0], C1[:,1], s=60, linewidth=1, c='y', label=t_1)
    axes.scatter(C2[:,0], C2[:,1], s=60, linewidth=1, c='b', label=t_2)
    axes.set_xlabel(lb_x)
    axes.set_ylabel(lb_y)
    return

#绘制三分类散点图
def plot3data(X, y, lb_x, lb_y, t_1, t_2, axes=None):
    c1 = np.in1d(y, [0])
    c2 = np.in1d(y, [1])
    c3 = np.in1d(y, [2])
    c1 = y == 0
    c2 = y == 1
    c3 = y == 2
    C1 = X[c1]
    C2 = X[c2]
    C3 = X[c3]
    if axes == None:
        axes = plt.gca()
    axes.scatter(C1[:,0], C1[:,1], s=60, linewidth=1, c='y', label=t_1)
    axes.scatter(C2[:,0], C2[:,1], s=60, linewidth=1, c='b', label=t_2)
    axes.scatter(C3[:,0], C3[:,1], s=60, linewidth=1, c='r', label=t_2)
    axes.set_xlabel(lb_x)
    axes.set_ylabel(lb_y)
    return

#绘制决策边界
def plotldb(model, X, axes=None):
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 0].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))
    X_G = np.c_[xx.ravel(), yy.ravel()]
    zz = model.predict(X_G)
    zz = zz.reshape(xx.shape)
    if axes == None:
        axes = plt.gca()
    axes.contour(xx, yy, zz, linwidth=1, color='r')
    color = ListedColormap(['#FFAAAA', '#AAFFAA'])
    axes.pcolormesh(xx, yy, zz, cmap=color)
    plotdata(X, y, "ES1", "ES2", "C1", "C2")

#绘制非线性决策边界
def plotcdb(model, X, poly,axes=None):
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 0].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))
    X_G = poly.fit_transform(np.c_[xx.ravel(), yy.ravel()])[:,1:]
    zz = model.predict(X_G)
    zz = zz.reshape(xx.shape)
    if axes == None:
        axes = plt.gca()
    axes.contour(xx, yy, zz, linwidth=1, color='r')
    color = ListedColormap(['#FFAAAA', '#AAFFAA'])
    axes.pcolormesh(xx, yy, zz, cmap=color)
    plotdata(X, y, "ES1", "ES2", "C1", "C2")
```

```python
#提取数据
X, y = make_moons(200, noise=0.2)
plotdata(X, y, "ES1", "ES2", "C1", "C2")
#LR分类器初始化
clf = LogisticRegressionCV()
#LR分类器训练
clf.fit(X, y)

#绘制线性决策边界
plotldb(clf, X)
#绘制非线性决策边界
poly = PolynomialFeatures(6)
X_C = poly.fit_transform(X)[:,1:]
clf.fit(X_C, y)
plotcdb(clf, X, poly)
```

## BP计算梯度以便使用(随机)梯度下降
![](http://s0.wp.com/latex.php?zoom=2&latex=%5Cbegin%7Baligned%7D++%26+%5Cdelta_3+%3D+%5Chat%7By%7D+-+y+%5C%5C++%26+%5Cdelta_2+%3D+%281+-+%5Ctanh%5E2+z_1%29+%5Ccirc+%5Cdelta_3W_2%5ET+%5C%5C++%26+%5Cfrac%7B%5Cpartial%7BL%7D%7D%7B%5Cpartial%7BW_2%7D%7D+%3D+a_1%5ET+%5Cdelta_3+%5C%5C++%26+%5Cfrac%7B%5Cpartial%7BL%7D%7D%7B%5Cpartial%7Bb_2%7D%7D+%3D+%5Cdelta_3%5C%5C++%26+%5Cfrac%7B%5Cpartial%7BL%7D%7D%7B%5Cpartial%7BW_1%7D%7D+%3D+x%5ET+%5Cdelta2%5C%5C++%26+%5Cfrac%7B%5Cpartial%7BL%7D%7D%7B%5Cpartial%7Bb_1%7D%7D+%3D+%5Cdelta2+%5C%5C++%5Cend%7Baligned%7D++&bg=ffffff&fg=000&s=0)

```python
#输出层的激活函数
def softmax(Z):
    h = np.exp(Z) / np.sum(np.exp(Z), axis=1, keepdims=True)
    return h
                           
#隐藏层的激活函数
def af(X):
    A = np.tanh(X)
    return A
                           
#损失函数                         
def costF(model, X, reg):
    #获得当前模型的权值矩阵和偏移矩阵
    W1,W2,b1,b2 = model['W1'],model['W2'],model['b1'],model['b2']
    #向前进行计算
    #计算Z1
    Z1 = X.dot(W1) + b1  #Z1为整个样本集向前计算的输入, b1为行向量
    #计算A1
    A1 = af(Z1)
    #计算Z2 
    Z2 = A1.dot(W2) + b2 #b2为行向量
    #计算A2
    A2 = softmax(Z2)
    y_bar = A2[range(X.shape[0]), y]
    #计算损失
    y_loss = (1.0 / X.shape[0]) * np.sum(-np.log(y_bar))
    r = (reg / 2.0) * (np.sum(np.square(W1)) + np.sum(np.square(W2)))
    J = y_loss + r
    #输出结果
    if np.isnan(J):
        return np.inf
    else:
        return J
    
#FC过程（正向计算过程)
def fc(model, X):
    #初始化参数                       
    W1,b1,W2,b2 = model['W1'], model['b1'], model['W2'], model['b2']
    #向前计算
    Z1 = X.dot(W1) + b1
    A1 = af(Z1)
    Z2 = A1.dot(W2) + b2
    A2 = softmax(Z2)
    return A1, A2
                           
#BP过程
def bp(model, X, A1, A2, reg=1, ep=0.01):
    #数据初始化                   
    W1, b1, W2, b2 = model['W1'], model['b1'], model['W2'], model['b2']
    
    #计算每一层的误差  delta3为输出层的误差， delta2为中间隐藏层的误差
    A2[range(X.shape[0]), y] -= 1
    delta3 = A2
    delta2 = (1 - np.power(A1, 2)) * (delta3.dot(W2.T))
    
    #计算权值和偏移的梯度
    dW2 = (A1.T).dot(delta3)
    db2 = np.sum(delta3, axis=0, keepdims=True)
    dW1 = (X.T).dot(delta2)
    db1 = np.sum(delta2, axis=0, keepdims=True)
    
    #权值和梯度加上正则化项
    dW2 += (reg * W2)
    dW1 += (reg * W1)
    
    #更新权值和偏置
    W2 += -ep * dW2
    b2 += -ep * db2
    W1 += -ep * dW1
    b1 += -ep * db1
    
    #更新model中的参数
    model ={'W1':W1, 'b1':b1, 'W2':W2, 'b2':b2}
    return 

#模型初始化
def init(inum, onum, hnum, nnum): 
    #初始化数据
    np.random.seed(0)
    W1 = np.random.randn(inum, nnum) / np.sqrt(inum)
    b1 = np.zeros((1, nnum))
    W2 = np.random.randn(nnum, onum) / np.sqrt(hnum)
    b2 = np.zeros((1, onum))                  
    #更新model中的属性值
    model ={'W1':W1, 'b1':b1, 'W2':W2, 'b2':b2}                 
    return model

#训练神经网络
def train(model, X, inum, onum, hnum=1):
    num_passes = 20000
    for i in xrange(0, num_passes):
        #正向计算
        A1, A2 = fc(model, X)
        #反向回归
        bp(model, X, A1, A2, 0.1, 0.01)
        if i % 1000 == 0: 
            print "Loss after iteration %i: %f" % (i, costF(model, X, 0.1))
            
#判定预测结果
def predict(model, X, y):
    #使用向前计算
    A1, A2 = fc(model, X)
    #得到预测值
    y_bar = np.argmax(A2, axis=1)
    #计算精度
    acc = 100.0 * np.sum(y_bar == y) / y.size
    return acc
        
#绘制非线性决策边界
def plotncdb(model, axes=None):
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 0].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))
    X_G = np.c_[xx.ravel(), yy.ravel()]
    #得到预测值
    zz = fc(model, X_G)[1]
    zz = np.argmax(zz, axis=1)
    zz = zz.reshape(xx.shape)
    #绘图
    if axes == None:
        axes = plt.gca()
    axes.contour(xx, yy, zz, linwidth=1, color='r')
    color = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    axes.pcolormesh(xx, yy, zz, cmap=color)
    plot3data(X, y, "ES1", "ES2", "C1", "C2")
    
```

```python
N = 100 # 每类样本个数
D = 2 # 维度
K = 3 # 类别数
X = np.zeros((N*K,D)) # data matrix (each row = single example)
y = np.zeros(N*K, dtype='uint8') # class labels
for j in xrange(K):
    ix = range(N*j,N*(j+1))
    r = np.linspace(0.0,1,N) # 半径
    t = np.linspace(j*4,(j+1)*4,N) + np.random.randn(N)*0.2 # 角度，加上是一个随机噪声
    X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
    y[ix] = j
# lets visualize the data:
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
```

```python
#初始化神经网络
model = init(2, 3, 1, 5)
#训练神经网络
train(model, X, 2, 3, 5)
#判定预测结果
acc = predict(model, X, y)
print "Accuracy {}%".format(np.round(acc, decimals=2))
#绘图
plotncdb(model, axes=None)
```

```python
# 然后听闻你想知道不同的隐层个数对结果的影响？
plt.figure(figsize=(16, 32))
hidden_layer_dimensions = [1, 2, 3, 4, 5, 20, 50, 60,70]
for i, nn_hdim in enumerate(hidden_layer_dimensions):
    plt.subplot(5, 2, i+1)
    plt.title('Hidden Layer size %d' % nn_hdim)
    #初始化神经网络
    model = init(2, 3, 1, nn_hdim)
    #训练神经网络
    train(model, X, 2, 3, nn_hdim)
    #绘图
    plotncdb(model, axes=None)
```
