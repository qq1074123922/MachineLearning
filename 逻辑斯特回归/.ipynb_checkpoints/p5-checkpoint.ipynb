```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.preprocessing import PolynomialFeatures
from scipy.optimize import minimize
%matplotlib inline
```

```python
def loaddata(file, dim):
    data = np.loadtxt(file, delimiter=dim)
    print data[:6,:6]
    return data

def plotdata(data, la_x, la_y, n_t, p_t, axes=None):
    pos = data[:,2] == 1
    neg = data[:,2] == 0
    pos_data = data[pos]
    neg_data = data[neg]
    if axes == None:
        axes = plt.gca()
    axes.scatter(pos_data[:,0], pos_data[:,1], marker='+', c='b', linewidth=2, label=p_t)
    axes.scatter(neg_data[:,0], neg_data[:,1], marker='o', c='y', linewidth=2, label=n_t)
    axes.set_xlabel(la_x)
    axes.set_ylabel(la_y)
    axes.legend(frameon=True, fancybox=True)

def sigmoid(z):
    s = 1.0 / (1.0 + np.exp(-z))
    return s

def cF(theta, X, y):
    theta = theta.reshape(-1,1)
    y = y.reshape(-1,1)
    m = y.size
    h = sigmoid(X.dot(theta))
    J = (-1.0 / m) * (np.log(h).T.dot(y) + np.log(1 - h).T.dot(1-y))
    if np.isnan(J[0]):
        return np.inf
    else:
        return J[0]

def g(theta, X, y):
    theta = theta.reshape(-1,1)
    y = y.reshape(-1,1)
    m = y.size
    h = sigmoid(X.dot(theta))
    grad = (1.0 / m) * X.T.dot(h - y)
    return grad.flatten()

def cFR(theta, X, y, R):
    theta = theta.reshape(-1, 1)
    y = y.reshape(-1, 1)
    m = y.size
    h = sigmoid(X.dot(theta))
    J = (-1.0 / m) * (np.log(h).T.dot(y) + np.log(1-h).T.dot(1-y)) + (1.0 * R / (2 * m)) * sum(np.square(theta))
    if np.isnan(J[0]):
        return np.inf
    else:
        return J[0]   
    
def gR(theta, X, y, R):
    theta = theta.reshape(-1,1)
    y = y.reshape(-1,1)
    m = y.size
    h = sigmoid(X.dot(theta))
    grad = (1.0 / m) * X.T.dot(h - y) + (1.0 * R / m) * theta
    return grad.flatten()

def predict(theta, X, y, threshold=0.5):
    theta = theta.reshape(-1, 1)
    y = y.reshape(-1,1)
    h = sigmoid(X.dot(theta)) >= 0.5
    h = h.astype("int")
    p = 100 * sum(h == y)[0] / y.size
    return p
```

```python
#非线性决策边界
data2 = loaddata("data2.txt",",")
#特征映射
poly = PolynomialFeatures(6)
X = poly.fit_transform(data2[:,:2])
theta = np.c_[np.zeros(X.shape[1])]
y = data2[:, 2]
#损失函数最小化(三种lambda)
fig, axes = plt.subplots(1, 3, sharey=True, figsize=(17,5))
for i, R in enumerate([0, 1, 100]):
    theta = np.c_[np.zeros(X.shape[1])]
    res2 = minimize(cFR, theta, args=(X, y, R), jac=gR, options={"maxiter":3000})
    #绘制非线性边界的数据初始化
    x1_min, x1_max = data2[:,0].min(), data2[:,0].max()
    x2_min, x2_max = data2[:,1].min(), data2[:,1].max()
    xx1, xx2 = np.meshgrid(np.linspace(x1_min, x1_max,100), np.linspace(x2_min, x2_max,100))
    X_P = np.c_[xx1.ravel(), xx2.ravel()]
    X_P_G = poly.fit_transform(X_P)
    h = sigmoid(X_P_G.dot(res2.x))
    h = h.reshape(xx1.shape)
    #绘制边界
    axes.flatten()[i].contour(xx1, xx2, h, [0.5], linewidth=1, colors='g')
    #绘制散点图
    plotdata(data2, "ES1", "ES2", "P", "F", axes.flatten()[i])
    #显示准确率
    acc = predict(res2.x, X, y)
    axes.flatten()[i].set_title("Accuracy {}% lambda = {}".format(np.round(acc, decimals=2), R))
    
```

```python

def fc(X,theta):
    Z1 = X.dot(theta)
    A1 = sigmoid(Z1)
    return A1

def bp(A1, X, y, theta, ep, reg):
    m = A1.size
    dtheta = (1.0 / m) * X.T.dot(A1 - y) + (1.0 * reg / m) * theta
    theta += -ep * dtheta
    
def init(m):
    theta = np.random.randn(m, 1)
    return theta

def train(theta, X, y, loops, ep=1, reg=1):
    y = y.reshape(-1,1)
    for i in xrange(0,loops):
        A1 = fc(X,theta)
        bp(A1, X, y, theta, ep, reg)
        if i % 1000 == 0:
            print costFR(theta, X, y, reg)  
            
def costFR(theta, X, y, reg):
    theta = theta.reshape(-1, 1)
    y = y.reshape(-1, 1)
    m = y.size
    h = sigmoid(X.dot(theta))
    J = (-1.0 / m) * (np.log(h).T.dot(y) + np.log(1-h).T.dot(1-y)) + (1.0 * reg / (2 * m)) * sum(np.square(theta))
    if np.isnan(J[0]):
        return np.inf
    else:
        return J[0]   
    
```

```python
from numpy import *

def loadData(name):
    data = []
    y = []
    f = open(name)
    for x in f.readlines():
        x = x.strip().split('\t')
        data.append(map(float, x))
        y.append(float(x[-1]))
    return array(data), array(y)

def fc(X, theta):
    Z1 = X.dot(theta)
    A1 = sigmoid(Z1)
    return A1

def bp(A1, y, theta, X, ep):
    m = A1.size
    dtheta = (1.0 / m) * X.T.dot(A1 - y)
    theta += -ep * dtheta
    return theta

def train(X, theta, y, ep=0.1, loops=1000):
    y = y.reshape(-1,1)
    for i in xrange(loops):
        A1 = fc(X, theta)
        theta = bp(A1, y, theta, X, ep)
        if i % 1000 ==0.0:
            costF(X, theta, y)        
            
def costF(X, theta, y):
    theta = theta.reshape(-1, 1)
    y = y.reshape(-1, 1)
    m = y.size
    h = sigmoid(X.dot(theta))
    J = (1.0 / m) * (log(h).T.dot(y) + log(1-h).T.dot(1-y))
    if isnan(J[0]):
        return inf
    else:
        return J[0]
```
