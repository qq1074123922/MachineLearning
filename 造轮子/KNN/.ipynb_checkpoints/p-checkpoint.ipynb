```python
from numpy import *
import operator
from os import listdir
```

```python
def KNN(x, X, y, k):
    E = sum(power(x - X, 2), axis=1)
    index = E.argsort()
    dictt = {}
    for i in xrange(k):
        label = y[index[i]]
        dictt[label] = dictt.get(label, 0) + 1
    result = sorted(dictt.iteritems(), key=operator.itemgetter(1), reverse=True)
    return result[0][0]
    
def f2m(name):
    f = open(name)
    content = f.readlines()
    m = len(content)
    X = zeros((m,3))
    y = []
    content = enumerate(content)
    for i, value in content:
        line = value.strip().split('\t')
        X[i,:] = line[:3]
        y.append(int(line[-1]))
    y = array(y)
    return X, y

def i2v(name):
    x = []
    f = open(name)
    for i in xrange(32):
        line = f.readline()
        for j in xrange(32):
            x.append(int(line[j]))
    return x

def formData(name):
    FL = listdir(name)
    m = len(FL)
    X = zeros((m, 1024))
    y = []
    for i in xrange(m):
        fname = FL[i]
        label = fname.split('.')[0].split('_')[0]
        y.append(label)
        X[i,:] = i2v('{}/{}'.format(name, fname))
    y = array(y)
    return X, y

def Norm(data):
    mi = data.min(0)
    ma = data.max(0)
    X = (data - mi) / (ma - mi)
    return X

def datingTest(name):
    X, y = f2m(name)
    X = Norm(X)
    m = X.shape[0]
    y_bar = []
    for i in xrange(m):
        x = X[i,:]
        result = KNN(x, X, y, 30)
        y_bar.append(result)

    y_bar = array(y_bar)
    print "Acc {}%".format(100.0 * sum(y_bar == y) / y.size)


def handwrittingTest(name1, name2):
    X1, y1 = formData(name1)
    X2, y2 = formData(name2)
    y_bar = []
    for i in xrange(X2.shape[0]):
        result = KNN(X2[i,:], X1, y1, 3)
        y_bar.append(result)
    y_bar = array(y_bar)
    acc = 100.0 * sum(y_bar == y2) / y2.size
    print "Acc is {}%".format(acc)
```
